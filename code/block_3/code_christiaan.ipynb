```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python


df = pd.read_csv('telnetscan.csv', parse_dates = ['timestamp'] )
```

```python
# df.ip.value_counts()
# df = df.drop_duplicates(['ip'])
df['timestamp'] = df['timestamp'].dt.date
```

```python
days = df.timestamp.unique()

```

```python
x = df.timestamp.value_counts().sort_index()
plt.plot(x)
```

```python
word_list = ["login", "username", "login" , "account", "user"]

result = list()
for day in days:
    vc = df.loc[df.timestamp == day].banner.value_counts()
    total_banner = 0
    total_count = 0
    for banner,count in vc.items():
        for word in word_list:
            if word in  banner.lower():
    #             print(banner)
                total_banner += 1 
                total_count += count
                break
    
    result.append(total_count/df.loc[df.timestamp == day].shape[0])
np.mean(result)
```

```python
# vc = df.banner.value_counts()

word_list_refuse = ["reject", "refused", "refuse", "rejected", 'block']

# total_banner_block = 0
# total_count_block = 0
# for banner,count in vc.items():
#     for word in word_list_refuse:
#         if (word in  banner.lower() or banner == ""):
# #             print(banner)
#             total_banner_block += 1 
#             total_count_block += count
#             break
# print(total_banner_block)
# print(total_count_block)

result2 = list()
for day in days:
    vc = df.loc[df.timestamp == day].banner.value_counts()
    total_banner_block = 0
    total_count_block = 0
    for banner,count in vc.items():
        for word in word_list_refuse:
            if(word in  banner.lower() or banner == ""):
    #             print(banner)
                total_banner_block += 1 
                total_count_block += count
                break
    
    result2.append(total_count_block/df.loc[df.timestamp == day].shape[0])
np.mean(result2)
```

```python
# df.info()


y = [np.mean(result), 1- np.mean(result)]
x = ["Authentication", "No authentication"]

Y = [np.mean(result2), 1 - np.mean(result2)]
X = ["Refuse", "Accept"]
```

```python
import matplotlib.pyplot as plt



bars = plt.bar(x, height=y, width=.4)

xlocs, xlabs = plt.xticks()

# reference x so you don't need to change the range each time x changes
xlocs=[i for i in x]
xlabs=[i for i in x]

plt.xlabel('Devices requiring authentication')

plt.xticks(xlocs, xlabs)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, yval)

plt.show()
```
